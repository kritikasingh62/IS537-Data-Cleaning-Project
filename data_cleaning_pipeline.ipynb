```python
county_ref = pd.read_csv(
    "national_county.txt",
    header=None,
    names=["state_abbr", "state_fips", "county_fips", "county_name", "class_code"],
    dtype=str
)

county_ref.head()
```

```python
def clean_county(value):
    """
    Standardize county/parish/borough text for joining.
    """
    if pd.isna(value):
        return np.nan

    value = str(value).upper().strip()

    # punctuation + separators
    value = value.replace("'", "")
    value = value.replace(".", "")
    value = value.replace("-", " ")

    # normalize common abbreviations
    value = value.replace(" E ", " EAST ")
    value = value.replace(" W ", " WEST ")
    value = value.replace(" N ", " NORTH ")
    value = value.replace(" S ", " SOUTH ")
    value = value.replace(" ST ", " SAINT ")
    value = value.replace(" STE ", " SAINTE ")

    # common typo cleanup
    value = value.replace("BOROUH", "BOROUGH")

    # remove geographic suffixes
    suffixes = [
        " COUNTY",
        " PARISH",
        " BOROUGH",
        " CENSUS AREA",
        " MUNICIPALITY",
        " CITY AND BOROUGH",
        " CITY",
        " MUNICIPIO",
    ]
    for suffix in suffixes:
        value = value.replace(suffix, "")

    # collapse spaces
    value = value.replace(" ", "")

    return value
```

```python
hgi["state_clean"] = hgi["State"].str.upper().str.strip()

hgi["county_clean"] = (
    hgi["County/Parish"]
    .str.upper()
    .str.strip()
    .str.replace(" COUNTY", "", regex=False)
    .str.replace(" PARISH", "", regex=False)
    .str.replace(" BOROUGH", "", regex=False)
    .str.replace(" CENSUS AREA", "", regex=False)
    .str.replace(" MUNICIPALITY", "", regex=False)
    .str.replace(" CITY AND BOROUGH", "", regex=False)
    .str.replace(" CITY", "", regex=False)
    .str.replace(".", "", regex=False)
    .str.replace("-", "", regex=False)
    .str.replace(" ", "", regex=False)
)
```

```python
# hgi["county_clean"] = hgi["county_clean"].str.replace("BOROUH", "BOROUGH", regex=False)
```

```python
hgi["state_clean"] = hgi["State"].astype(str).str.upper().str.strip()
hgi["county_clean"] = hgi["County/Parish"].apply(clean_county)

svi["state_clean"] = svi["ST_ABBR"].astype(str).str.upper().str.strip()
svi["county_clean"] = svi["COUNTY"].apply(clean_county)

county_ref["state_clean"] = county_ref["state_abbr"].astype(str).str.upper().str.strip()
county_ref["county_clean"] = county_ref["county_name"].apply(clean_county)
```

```python
county_fixes = {
    # District of Columbia
    ("DC", "THEDISTRICT"): "DISTRICTOFCOLUMBIA",

    # Alaska special geographies
    ("AK", "JUNEAU"): "JUNEAU",
    ("AK", "SITKA"): "SITKA",
    ("AK", "NORTHSLOPEBOROUGH"): "NORTHSLOPE",
    ("AK", "JUNEAU"): "JUNEAU",
    ("AK", "NORTHWESTARTIC"): "NORTHWESTARCTIC",
    ("AK", "SITKA"): "SITKA",

    # Louisiana
    ("LA", "EBATONROUGE"): "EASTBATONROUGE",
    ("LA", "JEFFRSONDAVIS"): "JEFFERSONDAVIS",
    ("LA", "STJOHNBAPTIST"): "STJOHNTHEBAPTIST",

    # Iowa / Minnesota / Nebraska
    ("IA", "OBRIEN"): "OBRIEN",
    ("MN", "YELLOWMEDCINE"): "YELLOWMEDICINE",
    ("MN", "LAKEOFWOODS"): "LAKEOFTHEWOODS",
    ("NE", "SCOTTBLUFF"): "SCOTTSBLUFF",

    # Puerto Rico
    ("PR", "SANJUAN"): "SANJUAN",
}
```

```python
def apply_county_fix(row):
    key = (row["state_clean"], row["county_clean"])
    return county_fixes.get(key, row["county_clean"])

hgi["county_clean"] = hgi.apply(apply_county_fix, axis=1)
```

```python
hgi_fips = pd.merge(
    hgi,
    county_ref[["state_clean", "county_clean", "state_fips", "county_fips", "county_name"]],
    on=["state_clean", "county_clean"],
    how="left"
)

hgi_fips["county_fips_full"] = hgi_fips["state_fips"] + hgi_fips["county_fips"]
hgi_fips.loc[hgi_fips["state_clean"] == "DC", "county_fips_full"] = "11001"
```

```python
total_hospitals = len(hgi_fips)
mapped_fips = hgi_fips["county_fips_full"].notna().sum()
fips_mapping_rate = mapped_fips / total_hospitals

print("Total hospitals:", total_hospitals)
print("Hospitals mapped to county FIPS:", mapped_fips)
print("New FIPS mapping rate:", mapped_fips / total_hospitals)
```

```python
leftovers = hgi_fips[hgi_fips["county_fips_full"].isna()][
    ["Facility Name", "County/Parish", "State", "county_clean"]
].copy()

print("\nRemaining unmatched after FIPS mapping:", len(leftovers))
print(leftovers.head(20))
```

```python
svi["FIPS"] = svi["FIPS"].astype(str).str.zfill(5)
```

```python
final_merged = pd.merge(
    hgi_fips,
    svi,
    left_on="county_fips_full",
    right_on="FIPS",
    how="left",
    suffixes=("_hgi", "_svi")
)

final_match_rate = final_merged["FIPS"].notna().mean()
print("\nFinal HGI-to-SVI join rate:", round(final_match_rate, 4))
```

```python
analysis_df = final_merged[
    final_merged["county_fips_full"].notna()
].copy()

print("\nAnalysis dataset shape:", analysis_df.shape)

print("\nRating summary:")
print(analysis_df["rating_clean"].value_counts(dropna=False).sort_index())
```

```python
if "RPL_THEMES" in analysis_df.columns:
    print("\nAverage SVI by hospital rating:")
    print(analysis_df.groupby("rating_clean")["RPL_THEMES"].mean())

    print("\nCorrelation between rating and SVI:")
    print(analysis_df[["rating_clean", "RPL_THEMES"]].corr())
else:
    print("\nColumn 'RPL_THEMES' not found. Check SVI column names.")
```

```python
analysis_df.to_csv("hospital_svi_final_merge.csv", index=False)
leftovers.to_csv("unmatched_after_cleaning.csv", index=False)
```

```python
summary = pd.DataFrame({
    "metric": [
        "HGI rows",
        "SVI rows",
        "Initial naive join match rate",
        "FIPS mapping rate",
        "Final HGI-to-SVI join rate",
        "Missing numeric ratings"
    ],
    "value": [
        len(hgi),
        len(svi),
        round(naive_match_rate, 4),
        round(fips_mapping_rate, 4),
        round(final_match_rate, 4),
        int(hgi["rating_clean"].isna().sum())
    ]
})
```
